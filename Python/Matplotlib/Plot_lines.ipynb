```python
#  A line in a diagram from position (0,0) to position (6,250):
import numpy as np
import matplotlib.pyplot as plt
xpoint=np.array([0,6])
ypoint=np.array([0,250])
plt.plot(xpoint,ypoint)
# The plot() function is used to draw points (markers) in a diagram.
plt.show()
```

```python
#A line in a diagram from position (0, 1) to position (-6,-250):
xpoint=np.array([0,-6])
ypoint=np.array([1,-250])
plt.plot(xpoint,ypoint)
 
plt.show()
```

```python
#Multiple Points
# just make sure you have the same number of points in both axis.
xpoints=[3,4,7,9]
ypoints=[0,4,2,0]
plt.plot(xpoints,ypoints)
plt.show()
```

```python
# Default X-Points
# If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points
ypoints=[3,87,10, 5, 7]
plt.plot(ypoints)
plt.show()
```

```python
# *linestyle*, or shorter *ls*, to change the style of the plotted line
xpoints=[3,45,56,23,2]
ypoints=[3,87,10, 5, 7]
plt.plot(xpoints,ypoints,ls='dotted')
#           OR
# plt.plot(xpoints,ypoints,ls=':')

plt.show()
```

```python
# color or the shorter c to set the color of the line
xpoints=[3,45,56,23,2]
ypoints=[3,87,10, 5, 7]
plt.plot(xpoints,ypoints,ls='--',color='r')
```

```python
# linewidth or the shorter lw to change the width of the line.
xpoints=[3,45,56]
ypoints=[3,87,10]
plt.plot(xpoints,ypoints,color='r',lw=12.7)
```

```python
#Multiple Lines
y1points=[3,45,56]
y2points=[3,87,56]
plt.plot(y1points,color='r')
plt.plot(y2points,color='hotpink')
plt.show()
```

```python
# To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'
xpoints=[0,6]
ypoints=[0,230]
plt.plot(xpoints,ypoints,'o')
plt.show()
 
```
