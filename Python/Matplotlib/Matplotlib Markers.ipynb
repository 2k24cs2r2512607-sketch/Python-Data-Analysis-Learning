```python
import matplotlib.pyplot as plt
```

```python
# Use the keyword argument "marker" to emphasize each point with a specified marker
ypoints = [2,4,6,8,23]
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# Mark each point with a star
ypoints = [2,4,6,8,23]
plt.plot(ypoints, marker = '*')
plt.show()
```

```python
# Use the keyword argument 'markersize' or the shorter version, 'ms' to set the size of the markers
xpoints=[3,4,5]
ypoints=[5,6,19]
plt.plot(xpoints,ypoints,'1:y',ms=20)
```

```python
# use the keyword argument 'markeredgecolor' or the shorter 'mec' to set the color of the edge of the markers
xpoints=[1,4,5,34]
ypoints=[5,6,19,12]
plt.plot(xpoints,ypoints,'o:b',ms=20,mec='m')
```

```python
# mfc to set the color inside the edge of the markers
xpoints=[1,4,5,34]
ypoints=[5,6,19,12]
plt.plot(xpoints,ypoints,'o:b',ms=20,mec='hotpink',mfc = '#4CAF50')
```
