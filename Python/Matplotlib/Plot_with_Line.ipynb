```python
#  A line in a diagram from position (0,0) to position (6,250):
import numpy as np
import matplotlib.pyplot as plt
xpoint=np.array([0,6])
ypoint=np.array([0,250])
plt.plot(xpoint,ypoint)
# The plot() function is used to draw points (markers) in a diagram.
plt.show()
```

```python
#A line in a diagram from position (0, 1) to position (-6,-250):
xpoint=np.array([0,-6])
ypoint=np.array([1,-250])
plt.plot(xpoint,ypoint)
 
plt.show()
```

```python
#Multiple Points
# just make sure you have the same number of points in both axis.
xpoints=[3,4,7,9]
ypoints=[0,4,2,0]
plt.plot(xpoints,ypoints)
plt.show()
```

```python
# Default X-Points
# If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points
ypoints=[3,87,10, 5, 7]
plt.plot(ypoints)
plt.show()
```
